```python
import pandas as pd
import random  # 랜덤한 초성을 선택하기 위해 random 모듈을 import합니다.
import time  # 게임 시간을 측정하기 위해 time 모듈을 import합니다.

# 데이터셋 로딩
df = pd.read_excel('초성게임 키워드.xlsx')

my_list = list(df.초성)  # 데이터프레임에서 초성 컬럼을 리스트로 가져옵니다.

max_tries = 3  # 문제를 맞출 때까지 최대 시도 횟수입니다.
score = 0  # 맞춘 문제 수를 기록하는 변수입니다.
num_questions = 0  # 출제된 문제 수를 기록하는 변수입니다.
start_time = time.time()  # 게임 시작 시간입니다.
game_time = 60  # 게임 시간(초)입니다.

# 게임 규칙을 출력합니다.
print("초성 게임을 시작합니다.\n")
print("\n게임 규칙\n")
print("pass를 입력하면 다음 문제로 넘어갑니다.")
print("end를 입력하면 게임이 강제 종료됩니다.")
print("answer를 입력하시면 정답을 확인할 수 있습니다.")
print("1분 안에 최대한 많은 문제를 맞춰보세요!\n")

# 사용자에게 "게임 시작" 문장을 입력받습니다.
while True:
    start_input = input('"게임시작"을 입력하면 게임이 시작됩니다.')
    if start_input == "게임시작":
        break
    else:
        print("잘못된 입력입니다. 다시 입력해주세요.\n")

# 게임 진행
start_time = time.time() # 게임 시작 시간입니다.

# 게임 진행
while True:
    # 게임 시간이 지나면 게임 종료
    if time.time() - start_time > game_time:
        print("\n게임 종료! 1분이 지났습니다.")
        print(f"맞춘문제수는 {num_questions}개 중 {score}개, 틀린문제수는 {num_questions-score}개입니다.")
        break
        
    # 랜덤 초성 선택
    random_value = random.choice(my_list)
    random_index = my_list.index(random_value)
    random_answer = df.한국어[random_index]
    random_keyword = df.분류[random_index]
    tries = 0
    
    # 문제 출력
    print(f"\n초성: {random_value}")
    print(f"분류: {random_keyword}")
    
    # 문제 풀기
    while tries < max_tries:
        answer = input("정답을 입력해주세요: ")
        
        # 정답인 경우
        if answer == random_answer:
            score += 1
            num_questions += 1
            print(f"\n정답입니다! 정답은 {random_answer}입니다.")
            print(f"현재 맞춘문제수는 {score}개입니다.")
            break
        # pass를 입력한 경우 다음 문제로 넘어감
        elif answer.lower() == "pass":
            num_questions += 1
            print("\n다음 문제로 넘어갑니다.")
            break
        # answer를 입력한 경우 정답 공개
        elif answer.lower() == "answer":
            print("\n정답:", random_answer)
            break
        # end를 입력한 경우 게임 강제 종료
        elif answer.lower() == "end":
            print("\n강제 종료.\n")
            game_time = 0
            break
        # 오답인 경우
        else:
            tries += 1
            num_questions += 1
            if tries == max_tries:
                print(f"\n틀렸습니다. 정답은 {random_answer}입니다. 다음 문제로 넘어갑니다.")
            else:
                print(f"\n틀렸습니다. {max_tries - tries}번 남았습니다.")
    
    # 남은 시간 출력
    print(f"남은 시간: {game_time - int(time.time() - start_time)}초")
```
